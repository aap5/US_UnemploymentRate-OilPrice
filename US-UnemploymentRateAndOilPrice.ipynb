```python
import quandl
import pandas as pd
```

```python
keys_file = open("apikey.txt")
lines = keys_file.readlines()
quandl.ApiConfig.api_key = lines[0].rstrip()
```

```python
oil = pd.DataFrame(quandl.get("FRED/DCOILWTICO", start_date = "2000-01-01", end_date = "2022-03-01"))
oil.head()
```

```python
import matplotlib.pyplot as plt
import datetime
```

```python
plt.figure()
plt.plot(oil, label='WTI Crude Oil')
plt.xlabel('date')
plt.ylabel('dollar per barrel')
plt.axhline(0, color = 'black')
plt.xlim(datetime.date(2007,8,1),datetime.date(2020,8,1))
plt.legend()
plt.title("WTI Crude Oil Price")
plt.savefig('oil.png')
```

```python
unemployment = pd.DataFrame(quandl.get("FRED/UNRATE", start_date = "2000-01-01", end_date = "2022-03-01"))
unemployment.head()
```

```python
fig, ax = plt.subplots(1,2)
ax1 = unemployment["Value"].plot(label = 'Unemployment Rate', figsize = (10,5), ax = ax[0], xlim = [pd.Timestamp('2000-01-01'), pd.Timestamp('2022-03-01')])

ax1.legend()
ax2 = oil["Value"].plot(label = "Crude Oil Price", ax = ax[1], xlim = [pd.Timestamp('2000-01-01'), pd.Timestamp('2022-03-01')])
ax2.legend(loc='upper left')
```
